## merge함수 : DataFrame합치기

```python
help(pd.merge)
```

```python
dfleft = DataFrame({'KEY': ['K0', 'K1', 'K2', 'K3'],
    'A': ['A0', 'A1', 'A2', 'A3'],
    'B': ['B0', 'B1', 'B2', 'B3']})


dfright = DataFrame({'KEY': ['K2', 'K3', 'K4', 'K5'],
    'C': ['C2', 'C3', 'C4', 'C5'],
    'D': ['D2', 'D3', 'D4', 'D5']})
```

```python
dfleft
```

```python
dfright
```

```python
#dfleft와 dfright를 Key변수를 기준으로 merge
pd.merge(dfleft,dfright, how="outer")
```

```python
pd.merge(dfleft,dfright)  #inner
pd.merge(dfleft,dfright, how="inner")
```

```python
pd.merge(dfleft,dfright, how="left") #left기준 (left에 있는 데이터프레임 모두 참조)
```

```python
pd.merge(dfleft,dfright, how="right") #right기준
```

```python
pd.merge(dfleft,dfright, how="right", on="KEY")
```

```python
pd.merge(dfleft,dfright, how="outer",on="KEY", indicator=True)
pd.merge(dfleft,dfright, how="outer",on="KEY", indicator='indicator_info')  
```

```python
dfleft2 = DataFrame({'KEY': ['K0', 'K1', 'K2', 'K3'],
     'A': ['A0', 'A1', 'A2', 'A3'],
     'B': ['B0', 'B1', 'B2', 'B3'],
     'C': ['C0', 'C1', 'C2', 'C3']})

dfright2 = DataFrame({'KEY': ['K0', 'K1', 'K2', 'K3'],
     'B': ['B0_2', 'B1_2', 'B2_2', 'B3_2'],
     'C': ['C0_2', 'C1_2', 'C2_2', 'C3_2'],
     'D': ['D0_2', 'D1_2', 'D2_2', 'D3_3']})
```

```python
pd.merge(dfleft2,dfright2, on="KEY", suffixes=('_left','_right')) # suffixes: 열 인덱스 이름 지정
```

## merge (index 기준) 

```python
#데이터프레임을 인덱스 기준으로 병합
dfleft = DataFrame({
    'A': ['A0', 'A1', 'A2', 'A3'],
    'B': ['B0', 'B1', 'B2', 'B3']}, index=['K0', 'K1', 'K2', 'K3'])


dfright = DataFrame({
    'C': ['C2', 'C3', 'C4', 'C5'],
    'D': ['D2', 'D3', 'D4', 'D5']}, index=['K2', 'K3', 'K4', 'K5'])
```

```python
dfleft
```

```python
dfright
```

```python
#index기준 병합 방법 : join(), merge()


#merge함수를 이용한 left join

#pd.merge(dfleft,dfright) #공통되는 열이 존재하지 않기 때문에 ERROR
pd.merge(dfleft,dfright, left_index=True, right_index=True, how="left")
```

```python
dfleft.join(dfright) #left기준 #how="left"
```

```python
#데이터프레임을 인덱스 기준으로 병합
dfleft = DataFrame({
    'A': ['A0', 'A1', 'A2', 'A3'],
    'B': ['B0', 'B1', 'B2', 'B3']}, index=['K0', 'K1', 'K2', 'K3'])


dfright = DataFrame({
    'C': ['C2', 'C3', 'C4', 'C5'],
    'D': ['D2', 'D3', 'D4', 'D5']}, index=['K2', 'K3', 'K4', 'K5'])
```

```python
pd.merge(dfleft,dfright,left_index=True, right_index=True, how="left")
```

```python
dfleft.join(dfright)
```

```python
pd.merge(dfleft,dfright,left_index=True, right_index=True, how="outer")
```

```python
dfleft.join(dfright, how="outer")
```

## merge(column + index 기준) 

```python
dfleft = DataFrame({
    'KEY':['K0', 'K1', 'K2', 'K3'],
    'A': ['A0', 'A1', 'A2', 'A3'],
    'B': ['B0', 'B1', 'B2', 'B3']})


dfright = DataFrame({
    'C': ['C2', 'C3', 'C4', 'C5'],
    'D': ['D2', 'D3', 'D4', 'D5']}, index=['K2', 'K3', 'K4', 'K5'])
```

```python
pd.merge(dfleft,dfright,left_on='KEY', right_index=True, how="left")
```

```python
dfleft.join(dfright, on="KEY", how="left")
```
