```python
#subplot(2,1,1) #2헹1열의 위쪽
#subplot(2,1,2) #2행 1열의 아랫쪽

x1= np.linspace(0.0,5.0)
x1

x2= np.linspace(0.0,2.0)
x2

y1=np.cos(2*np.pi*x1)*np.exp(-x1)
y2=np.cos(2*np.pi*x2)

ax1=plt.subplot(2,2,1)
plt.plot(x1,y1,"yo-")   #노랑선, 포인트점
plt.title("table1")

ax2=plt.subplot(2,1,2)
plt.plot(x2,y2,"r-")   #빨간선
plt.title("table2")

plt.tight_layout() 

```

```python
d1=anscombe[anscombe["dataset"]=="I"]
d2=anscombe[anscombe["dataset"]=="II"]
d3=anscombe[anscombe["dataset"]=="III"]
d4=anscombe[anscombe["dataset"]=="IV"]
plt.plot(d1["x"],d1["y"]
        
        )
```

```python
fig=plt.figure()
axes1=fig.add_subplot(2,2,1)
axes2=fig.add_subplot(2,2,2)
axes3=fig.add_subplot(2,2,3)
axes4=fig.add_subplot(2,2,4)

axes1.plot(d1["x"],d1["y"],"o")
axes2.plot(d2["x"],d2["y"],"o")
axes3.plot(d3["x"],d3["y"],"o")
axes4.plot(d4["x"],d4["y"],"o")

axes1.set_title("dataset1")
axes2.set_title("dataset2")
axes3.set_title("dataset3")
axes4.set_title("dataset4")

fig.tight_layout()     #겹치는 부분 fit하게 설정
```

```python
tips=sns.load_dataset("tips")
tips
```

```python
fig=plt.figure()
axes1=fig.add_subplot()


scPlot=plt.figure()

axes1.hist(tips["total_bill"])
axes1=scPlot.add_subplot(1,1,1)
axes1.scatter(tips["total_bill"],tips["tip"]) #x="total_bill", y="tip"
```

```python
tips["total_bill"]
```

```python
#boxPlot(상자그림)
boxPlot=plt.figure()

axes1=boxPlot.add_subplot(1,1,1)
axes1.boxplot([tips[tips["sex"]=="Female"]["tip"],
tips[tips["sex"]=="Male"]["tip"]],
labels=["Female","Male"])
```

```python
tips["sex_color"]=np.where(tips["sex"]=="Female",0,1)
tips
```

```python
sPlot=plt.figure()
axes1=sPlot.add_subplot(1,1,1)
axes1.scatter(x=tips["total_bill"], y=tips["tip"], s=tips["size"]*10, c=tips["sex_color"], alpha=0.5) #alpha(투명도)
plt.show()
```

```python
#distplot : seaborn에서 히스토그램을 그릴때 사용

ax=plt.subplots()
sns.distplot(tips["total_bill"], kde=False)  #kde 옵션 : 밀집도 
```

```python
ax=sns.distplot(tips["total_bill"], rug=True)  #rug옵션: x축상의 밀도
```
