```python
"""
titanic 데이터로 Kaggle등수 올리기(모델 만들기)

-조건문 개선
-Name => Age 결측값 처리
(Mrs => Mrs의 중앙값...)
-FamilySize 컬럼 추가
(SibSp+Parch)

"""
```

```python
import pandas as pd
import numpy as np
```

```python
train=pd.read_csv("train.csv")
test=pd.read_csv("test.csv")
```

```python
#2번 Name => Age 결측값 처리
train["title"]=train["Name"].str.extract(" ([A-Za-z]+)\.")
```

```python
train["title"].value_counts() #각 title별 요소 값

#train["Age"].isnull().value_counts()  #Age열의 각 null값 

#train.groupby("title")["Age"].describe() #title별 Age 요약통계
```

```python
train[train["title"]=="Mrs"].isnull().sum()
train[train["title"]=="Mrs"].median()["Age"] #35.0 추출

train[train["title"]=="Mrs"].median()["Age"]
```

```python
#train[train["title"]=="Mr"].median()   #Mr의 Age 중앙값 #30
#Mrs_m=train[train["title"]=="Mrs"].median() #Mrs의 Age 중앙값 #35
#Miss_m=train[train["title"]=="Miss"].median() # Miss의 Age중앙값  #21
#Mast_m=train[train["title"]=="Master"].median() #Master의 Age 중앙값 #3.5
train[train["title"]=="Mrs"].median()["Age"]
```

```python
#train["Age"]=train.loc[train["title"]=="Mrs","Age"].fillna(train[train["title"]=="Mrs"].median()["Age"])  #실패/ 안되는이유 모름ㅠ
#train["Age"].isnull().sum()
```

```python
train["Age"]=train["Age"].fillna(train.groupby("title")["Age"].transform("median")) 
#각 title별 중앙값을 각title별 Age열의 NA에 대체
```

```python
Age_csv=train[["Age"]]
Age_csv.to_csv("Age_csv.csv",index=False)
```

```python
#3번 FamilySize 컬럼 추가 (SibSp+Parch)


train["FamilySize"]=train["SibSp"]+train["Parch"]
train
```

```python
bins=np.linspace(train["FamilySize"].min()-1,train["FamilySize"].max()+1,4)

bins   #구간 보기
```

```python
labels=["Small","Medium","Large"]
train["Family_bin"]=pd.cut(train["FamilySize"], bins, labels=labels)

train
```

```python
test=pd.read_csv("test.csv")
test["title"]=test["Name"].str.extract(" ([A-Za-z]+)\.")
```

```python
test["Age"]=test["Age"].fillna(test.groupby("title")["Age"].transform("median")) 
```

```python
test["FamilySize"]=test["SibSp"]+test["Parch"]

bins=np.linspace(test["FamilySize"].min()-1,test["FamilySize"].max()+1,4)

labels=["Small","Medium","Large"]
test["Family_bin"]=pd.cut(test["FamilySize"], bins, labels=labels)
test
```

```python
test["Survived"]=(test.Sex=="female") & (test.Embarked) & (test.Age>0)
test["Survived"].value_counts()
```

```python
submission2=test[["PassengerId","Survived"]]
submission2["Survived"]=submission2["Survived"].astype(int)
submission2.head(10)

submission2.to_csv("submission2.csv", index=False) #csv저장
```
