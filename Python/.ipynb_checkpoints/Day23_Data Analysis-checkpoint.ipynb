## ** 리스트/ 문자열의 "메모리주소할당" 차이점

```python
a=[1,2,3]
id(a) #변수의 메모리 주소값

a[1]=20
id(a)  #리스트 안의 값을 변경해도 주소는 변경안됨

b={1,2,3}
id(b)
```

```python
s="abc"
#print(id(s))
s="kbc"
s
print(id(s)) #주소위치 변경됨
```

```python
a=[1,2,3]
b=a  #얕은복사 : 값이 할당되는 것이 아니라 메모리 주소를 공유
print(id(a))
print(id(b))
```

```python
s="abc"

v=s
print(id(s))
print(id(v))

v="edf"
print(id(s))
print(id(v))  #문자열의 경우도 리스트처럼 동일한 메모리 주소를 공유하게 되지만 새로운 값이 부여되면
               #새로운 메모리 주소를 갖게됨.
```

```python
a=[1,2,3]
b=a
print(id(a))
print(id(b))
print(a is b) #메모리상의 주소가 동일한지 비교
print(a==b) #값이 동일한지 비교
```

```python
a=[1,2,3]
b=a[:]
print(id(a))
print(id(b)) #메모리 주소가 다름   
print(a is b) #False
print(a==b) 
```

```python
#windspeed열 안 0값 대체 => 상관관계를 따져서 대체해주어야 함(정석)
#windspeed열 안 0값 => 평균 

#train.loc[train["windspeed"]==0,"windspeed"]=train["windspeed"].mean()

trainWind0=train.loc[train["windspeed"]==0]
trainWindNot0=train.loc[train["windspeed"]!=0]

trainWind0=test.loc[test["windspeed"]==0]
trainWindNot0=test.loc[test["windspeed"]!=0]

print(trainWind0.shape)
print(trainWindNot0.shape)
```

```python
#머신러닝 랜덤포레스트로 풍속예측

from sklearn.ensemble import RandomForestClassifier

def predict_windspeed(data):
    #data의 windspeed값이 0인 데이터를 랜덤포레스트를 이용하여 예측한 값으로 대체
    
    #풍속 예측에 사용되는 변수 (임시) => 상관분석을 해야됨(정석)
    wCol=["season","weather","humidity","month","temp","year","atemp"]
    
    dataWind0=data.loc[data["windspeed"]==0]
    dataWindNot0=data.loc[data["windspeed"]!=0]
    
    #랜덤포레스트 분류기 생성
    
    rfModel=RandomForestClassifier()
    
    #wCol변수에 따라 풍속값이 일반화됨
    #wCol => 풍속 
    #fit()함수를 사용하려면 windspeed의 타입 변형 (int->stringd)
    
    dataWindNot0["windspeed"]=dataWindNot0["windspeed"].astype(str)
    
    rfModel.fit(dataWindNot0[wCol],dataWindNot0["windspeed"]) 
    
    #학습모델로 풍속 0에 대한 데이터를 예측
    preValue=rfModel.predict(dataWind0[wCol])
    
    predictWind0=dataWind0
    predictWindNot0=dataWindNot0
    
    predictWind0["windspeed"]=preValue
    
    data=predictWindNot0.append(predictWind0) 
#     data.reset_index(inplace=True)
#     data.drop("index", inplace=True, axis=1)
    return data
```

```python
#feature selection
#(???)=>count?
#연속형(temp, humi, wind, atime)
#범주형 변수는 타입을 category로 변경

feature_names=(['season', 'holiday', 'workingday', 'weather', 'temp',
       'atemp', 'humidity', 'windspeed', 'year', 'hour', 'dayofweek'])

c_f_n=(['season', 'holiday', 'workingday', 'weather', 'year', 'hour','month','dayofweek'])

for v in c_f_n:  
    train[v]=train[v].astype("category")
    test[v]=test[v].astype("category")
```

```python
xtrain=train[feature_names]
xtest=test[feature_names]

xtrain.shape
xtest.shape
```

```python
ytrain=train["count"] #레이블(정답)
ytrain.shape
```

```python
def rmsle(predicted_value, actual_value):
    #넘파이 배열로 변환
    predicted_value=np.array(predicted_value)
    actual_value=np.array(actual_value)
    
    log_predict=np.log(predicted_value+1)
    log_actual=np.log(actual_value+1)
    
    diff=log_predict-log_actual
    diff=np.square(diff) #diff**2 (제곱)
    
    mean_diff=diff.mean()
    score=np.sqrt(mean_diff)
    return score


from sklearn.metrics import make_scorer
rmsle_scorer=make_scorer(rmsle)
rmsle_scorer
```

## cross validation(교차검증)

```python
#데이터의 폴드(train/test)를 바꿔가면서 검증
# data : 400 
#fold: 1 2 3 4 (k=4)
#train :1 2 3 / test : 4 => 모델평가 80%
#       1 2 4 / test : 3  => 70%
#       1 3 4 / test: 2 => 80%
#       2 3 4 / test : 1 =>70%
```

```python
from sklearn.model_selection import KFold
from sklearn.model_selection import cross_val_score
```

```python
kfold=KFold(n_splits=10, shuffle= True, random_state=42)
```

```python
from sklearn.ensemble import RandomForestRegressor
model=RandomForestRegressor(n_estimators=100, n_jobs=-1, random_state=42) #n_jobs(코어성능): -1(최적화)
```

```python
%time score=cross_val_score(model, xtrain, ytrain, cv=kfold,scoring=rmsle_scorer)
print(score.mean())
```

```python
model.fit(xtrain,ytrain)

prediction=model.predict(xtest)
prediction
```
