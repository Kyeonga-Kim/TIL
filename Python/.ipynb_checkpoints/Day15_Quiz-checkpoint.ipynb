```python
#Q1-1. df1~df3 merge하기(위아래)
```

```python
import pandas as pd

df1=pd.read_csv("concat_1.csv")  #df1 데이터 읽어오기
df1
```

```python
df2=pd.read_csv("concat_2.csv")  #df2 데이터 읽어오기
df2
```

```python
df3=pd.read_csv("concat_3.csv")   #df3 데이터 읽어오기
df3
```

```python
df123=pd.concat([df1,df2,df3])  #위아래로 합치기 
df123
```

```python
#Q1-2. axis옵션, how옵션
```

```python
df123_a1=pd.concat([df1,df2,df3], axis=1)  #axis=1옵션: 좌우로 합치기 
df123_a1
```

```python
df123_o=pd.concat([df1,df2,df3], join="outer") #join="outer"옵션
df123_o
```

```python
#Q1-3.ignore_index적용
```

```python
df123_ig=pd.concat([df1,df2,df3],ignore_index=True )
df123_ig
```

```python
#Q1-4. merge/join 함수버전 각각 사용
```

```python
#merge 함수
df12=pd.merge(df1,df2, how="outer")  
df12
```

```python
df123=pd.merge(df12,df3, how="outer")  
df123
```

```python
#join함수
df12.join(df3, how="left", rsuffix=("2"))          
```

```python
Q3. ppt 알고리즘 문제
-왼쪽과 오른쪽 창문을 열었을 때 양쪽(좌,우)로 모두 2칸이상의 공간이 확보되는 세대의 수를 반환
-빌딩 높이는 0-255
-초기값 : [0,0,3,5,2,4,9,0,6,4,0,6,0,0]
```

```python
building=[0,0,3,5,2,4,9,0,6,4,0,6,0,0] #빌딩의 초기 높이값
i=0
res=0
for i in range(2,len(building)-2):   #building 리스트의 0값을 제외한 범위부터 시작
    compare=building[i-2:i+3]       # i-2부터 i+2까지 총 5개 요소를 compare에 넣기
    compare.remove(building[i])     #중간 자리 인덱스 i를 compare에서 제거
    m=max(compare)                  #compare에 최대값을 m
    if building[i]>m:              #building[i]와 building[i]를 제거한 compare에서의 최댓값 비교
        res+=building[i]-m         #building[i]-m 차이값 저장
print(res)
            
```
