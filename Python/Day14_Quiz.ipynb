```python
zoo_name=[]

f=open("C:/Users/student/Downloads/zoo.names","r")
for line in f:
    pat=re.compile("\s+[0-9]+[.]\s+(?P<name>[ a-z]+)")
    m=pat.findall(line)
    #print(m)
    if m!=[]:
        zoo_name.append(m[0])
print(zoo_name)
               
f.close()
```

```python
zoo=pd.read_csv("zoo.data", names=zoo_name, index_col="animal name")
zoo
```
