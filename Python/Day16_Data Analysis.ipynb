## isnull, notnull 함수

```python
# isnull, notnull함수
pd.isnull(NaN)
```

```python
df_left=pd.DataFrame({
    "KEY" : ['K0','K1','K2','K3'],
    "A" : ['A0','A1','A2','A3'],
    "B" : [0.5, 1.0, 0.7, 1.5]
})
df_right=pd.DataFrame({
    "KEY" : ['K2','K3','K4','K5'],
    "C" : ['C2','C3','C3','C5'],
    "D" : ['D2','D3','D3','D5']
})
```

```python
df_left
```

```python
df_right
```

```python
df_all=pd.merge(df_left, df_right, on="KEY", how="outer") 
df_all
```

```python
pd.isnull(df_all)

pd.notnull(df_all) #null값이 아닌 값에 대해 True

df_all.isnull() #null값에 대해 True
```

```python
df_all.ix[[0,1],['A','B']]=None  
df_all
```

```python
df_all[['A','B']].isnull()  #None도 Null값으로 인식
```

## isnull, notnull 합계(갯수) 구하기 

```python
df_all.isnull().sum() #컬럼별 결측값 갯수

#A열의 결측값 개수 출력
df_all['A'].isnull().sum() #True인 값 총 4개 => 4 출력
```

```python
#각 열별로 결측값이 아닌 개수 출력
df_all=df_all.notnull().sum()
```

```python
df_all.notnull().sum(axis=1) #가로로 갯수 카운트
```

```python
df_all['Nan_cnt']=df_all.isnull().sum(axis=1) #옆으로 열을 합쳐야 되니깐 axis=1
df_all['NotNull_cnt']=df_all.notnull().sum(1) #axis 생략가능 

df_all
```

## NaN에 대체값 넣기

```python
df=pd.DataFrame(np.random.randn(5,3),columns=['c1','c2','c3'])   #표준정규분포를 따르는 난수 발생
df

df.ix[0,0]=None
df


df.ix[1,['c1','c3']]=None
df.ix[2,['c2']]=None
df.ix[3,['c2']]=None
df.ix[4,['c3']]=None
df
```

```python
df.fillna(0)   #NaN자리를 모두 0으로 출력
```

```python
df_miss=df.fillna("missing")  #NaN자리를 모두 "missing"으로 출력
df_miss
```

```python
df.fillna(method="ffill", limit=1)  #ffill옵션 : 아래->위로 채우기  #limit옵션 : n개만 채우기
```

```python
df
```

```python
df.mean()   #각 columns값들의 평균, NaN은 평균에서 제외
```

```python
# NaN자리를 평균값으로 채우기

df.fillna(df.mean())
```
