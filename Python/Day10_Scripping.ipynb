## find 함수 

```python
#find 함수: id를 이용하여 직접 접근

html="""
<html><body>
<h1 id="title">스크래핑</h1>
<p id="body">웹 페이지 분석</p>
<p>원하는 부분 추출</p>
</body></html>
"""

soup=BeautifulSoup(html, 'html.parser') 

title=soup.find(id="title")
body=soup.find(id="body")
print("title: " +title.string)
print("body: " +body.string)
```

## find_all 함수 

```python
#find_all: 여러개의 태그를 한번에 추출

html="""
<html>
<body>
<ul>                   #<ol>:순서가 있다 <ul>:순서가 없다
<li><a href="http://www.naver.com">naver</a></li>   #링크 태그
<li><a href="http://www.daum.com">daum</a></li>
</ul>
</body>
</html>
"""
```

```python
soup=BeautifulSoup(html, "html.parser")
links=soup.find_all("a") # a태그에 해당하는 데이터 모두 출력

for a in links:
    if "href" in a.attrs: 
    href=a.attrs['href']  #a 속성 추출
    text=a.string
    print(text,":", href)
    #print(href)
    
soup
```

```python
html="""
<html>
    <body>
        <div id="myid">
            <h1>2020년</h1>
            <ul class='day'>
                <li>월요일</li>
                <li>화요일</li>
                <li>수요일</li>
            </ul>
        </div>
    </body>
</html>
"""
```

```python
soup=BeautifulSoup(html, "html.parser")
soup
```

```python
soup.select_one("div#myid h1")
soup.select_one("div#myid > h1") #myid태그 안의 h1태그를 출력
print(soup.select_one("div#myid h1").string) 
```

```python
print(soup.select_one("div#myid ul")) 

print(soup.select_one("div#myid > ul.day > li"))  #class 출력: ul.class이름 #select_one: 요소 하나만

myList=soup.select("div#myid > ul.day > li")

for a in myList:
    print(a.string)
```
