```python
#2번

live=titanic.loc[titanic["Survived"]==1] #Survived열:1 =>생존
dead=titanic.loc[titanic["Survived"]==0]  #Survived열:0 => 사망

bins = [5, 20, 30, 40, 60, 90]
labels = ["미성년자", "청년", "중년", "장년", "노년"]

titanic["age_cat"]=pd.cut(titanic["Age"],bins,labels=labels) #주어진 bins(구간) 과 label을 가지고 Age에 대해 분류

live_percent=live["age_cat"].value_counts()/live["age_cat"].value_counts().sum() #생존자 각 그룹별 승객 비율
dead_percent=dead["age_cat"].value_counts()/dead["age_cat"].value_counts().sum() #사망자 각 그룹별 승객 비율
```

```python
#1-3번 
from sklearn.preprocessing import 

1)표준화 : (각데이터-평균) /표준편차
sklearn의 표준화 함수 
-StandardScaler() : 평균(0)과 표준편차(1)이 되도록 표준화
-RobustScaler() : 중앙값(0)과 IQR(Q3-Q1)(1)이 되도록 표준화    
  

2)정규화 : 0~1사이의 범위로 데이터를 표준화
sklearn의 정규화 함수
-MinMaxScaler(): 최소최대 0~1 범위로 정규화
-minmax_scale(): 최소최대 0~1 범위로 정규화
```

```python
#2번.
표준화의 수식은 (각데이터-평균) /표준편차이며, 모집단이 정규분포를 따르는 경우에 평균을 0 , 표준편차가 1인 표준 정규 분포로 표준화하는 방법이다.
표준화를 구하는 방법은 크게 numpy(수식)방법 , scipy-stats-zscore()로 표준화하는 방법, sklearn-StandardScaler()방법이 있다.
추가로 sklearn-RobustScaler()함수로는 중앙값과 IQR으로 표준화할수 있다.

정규화는 0~1사이의 범위로 데이터를 표준화시키며, 수식은 (각 데이터 값-최솟값) / (최대값-최소값)이다.
정규화를 구하는 방법은 크게 수식으로 구하는 방법 , sklearn-MinMaxScaler()으로 구하는 방법, sklearn-minmax_scale()으로 구하는 방법이 있다.
```

```python
[파이썬 기본문법과 통계기초]
다음 실행 예시처럼 교환할 돈을 입력받아서, 최소한의 동전의 개수로 교환해 주는 파이썬 코드를 작성하시오.
바꿀 돈 -->7777
500원: 15 , 100원: 2 , 50원: 1 , 10원 2 , 나머지: 7                
```

```python
money=7777

while True:
    ohbak=int(money/500)            #몫 = 500원 갯수
    money=money%500
    bak=int(money/100)              #몫 = 100원 갯수
    money=money%100
    ohship=int(money/50)            #몫 = 50원 갯수
    money=money%50
    ship=int(money/10)              #몫 = 10원 갯수
    res=money%10                   #나머지 = 나머지 수
    
    if res<10:              # 나머지가 10보다 작으면 while문 중지
        break
        
print("500원: %d , 100원: %d, 50원: %d, 10원: %d, 나머지: %d" %(ohbak,bak,ohship,ship,res))
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

train=pd.read_csv("train.csv", parse_dates=["datetime"])
train
```

```python
train["hour"]=train["datetime"].dt.hour
train["dayofweek"]=train["datetime"].dt.dayofweek


fig,(ax1,ax2)=plt.subplots(nrows=2)
fig.set_size_inches(15,8)
sns.pointplot(data=train, x="hour",y="count",hue="workingday",ax=ax1) #workingday
sns.pointplot(data=train, x="hour",y="count",hue="dayofweek",ax=ax2)  #dayofweek
```
