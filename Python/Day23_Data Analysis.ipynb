## ** 리스트/ 문자열의 "메모리주소할당" 차이점

```python
a=[1,2,3]
id(a) #변수의 메모리 주소값

a[1]=20
id(a)  #리스트 안의 값을 변경해도 주소는 변경안됨

b={1,2,3}
id(b)
```

```python
s="abc"
#print(id(s))
s="kbc"
s
print(id(s)) #주소위치 변경됨
```

```python
a=[1,2,3]
b=a  #얕은복사 : 값이 할당되는 것이 아니라 메모리 주소를 공유
print(id(a))
print(id(b))
```

```python
s="abc"

v=s
print(id(s))
print(id(v))

v="edf"
print(id(s))
print(id(v))  #문자열의 경우도 리스트처럼 동일한 메모리 주소를 공유하게 되지만 새로운 값이 부여되면
               #새로운 메모리 주소를 갖게됨.
```

```python
a=[1,2,3]
b=a
print(id(a))
print(id(b))
print(a is b) #메모리상의 주소가 동일한지 비교
print(a==b) #값이 동일한지 비교
```

```python
a=[1,2,3]
b=a[:]
print(id(a))
print(id(b)) #메모리 주소가 다름   
print(a is b) #False
print(a==b) 
```

```python
a=[[1,2],[3,4]]
print(id(a))
print(id(a[0]))
print(id(a[1]))


b=a[:]
print(id(b))
print(id(b[0])) #a[0]와 동일한 주소
print(id(b[1])) #a[1]와 동일한 주소

a[0]=[100,200]
a
b
print(id(a[0]))
```

```python
#깊은 복사
#내부 객체들이 서로복사

import copy
a=[[1,2],[3,4]]
b=a
print(id(a))
print(id(b))

a[1]=[5,6]
a
b


a=[[1,2],[3,4]]
b=copy.deepcopy(a)
a[1].append(5)
print(id(a))
print(id(b))

print(a)
print(b)
```

```python
#outlier 제거


(train["count"].std()*3) #표준편차 3배

trainWithoutOutlier=train[np.abs(train["count"]-train["count"].mean()) <= (train["count"].std()*3)]

print(train.shape)
print(trainWithoutOutlier.shape)
```

```python
#windspeed열 안 0값 대체 => 상관관계를 따져서 대체해주어야 함(정석)
#windspeed열 안 0값 => 평균 

#train.loc[train["windspeed"]==0,"windspeed"]=train["windspeed"].mean()

trainWind0=train.loc[train["windspeed"]==0]
trainWindNot0=train.loc[train["windspeed"]!=0]

print(trainWind0.shape)
print(trainWindNot0.shape)
```

```python
#머신러닝 랜덤포레스트로 풍속예측

from sklearn.ensemble import RandomForestClassifier

def predict_windspeed(data):
    #data의 windspeed값이 0인 데이터를 랜덤포레스트를 이용하여 예측한 값으로 대체
    
    #풍속 예측에 사용되는 변수 (임시) => 상관분석을 해야됨(정석)
    wCol=["season","weather","humidity","month","temp","year","atemp"]
    
    dataWind0=data.loc[data["windspeed"]==0]
    dataWindNot0=data.loc[data["windspeed"]!=0]
    
    #랜덤포레스트 분류기 생성
    
    rfModel=RandomForestClassifier()
    
    #wCol변수에 따라 풍속값이 일반화됨
    #wCol => 풍속 
    #fit()함수를 사용하려면 windspeed의 타입 변형 (int->stringd)
    
    dataWindNot0["windspeed"]=dataWindNot0["windspeed"].astype(str)
    
    rfModel.fit(dataWindNot0[wCol],dataWindNot0["windspeed"]) 
    
    #학습모델로 풍속 0에 대한 데이터를 예측
    preValue=rfModel.predict(dataWind0[wCol])
    
    predictWind0=dataWind0
    predictWindNot0=dataWindNot0
    
    predictWind0["windspeed"]=preValue
    
    data=predictWindNot0.append(predictWind0)    
    return data

predict_windspeed(train)

```
