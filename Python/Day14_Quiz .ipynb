```python
import pandas as pd 
import numpy as np
from pandas import DataFrame 

DataFrame()

import re
```

```python
zoo_name=[]     #이름 넣을 빈 리스트 생성

f=open("C:/Users/Kim/Downloads/zoo.names","r")    #zoo.name 파일 읽기
for line in f:
    pat=re.compile("\s+[0-9]+[.]\s+(?P<name>[ a-z]+)") #각 라인을 정규표현식을 이용해 추출
    m=pat.findall(line)
    if m!=[]:           # 빈 리스트가 아닌 이름이 들어있는 리스트면 zoo_name 리스트에 넣기
        zoo_name.append(m[0])  
print(zoo_name) 
               
f.close()
```

```python
zoo=pd.read_csv("zoo.data", names=zoo_name, index_col="animal name")
zoo
```

```python
Q2.PPT문제 

-제한된 횟수만큼 평탄화(덤프) 작업을 한 이후, 최고점과 최저점의 차이를 리턴하는 함수를 구현하시오
-가장 높은 곳에 있는 상자를 가장 낮은곳에 옮기는것을 덤프라고 한다.
-아래의 예시에서 1회 덤프 후, A 상자가 B로 이동하였다.

-상자의 높이 : 1~100 
-덤프횟수: 10번
```

```python
box = [5,8,3,1,4,6,9,9,2,2,4]
times = 10                  #횟수: 10번

for i in range(times):   #횟수 10번 반복
    
    max_i=box.index(max(box)) #box의 max값의 index를 max_i에 저장
    min_i=box.index(min(box)) #box의 min값의 index를 min_i에 저장

    box[max_i]-=1   #max_i인덱스의 값 -1
    box[min_i]+=1   #min_i 인덱스의 값 +1
    
print(max(box)-min(box)) #box리스트의 최댓값 최솟값 차이
```
