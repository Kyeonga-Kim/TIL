```python
#1-1)Pclass를 원핫인코딩
import pandas as pd
import numpy as np
from sklearn.preprocessing import OneHotEncoder

titanic=pd.read_csv("train .csv")
pd.get_dummies(titanic["Pclass"]) #Pclass의 이산화 변수를 이항변수화 => 원핫인코딩
```

```python
#1-2) Age를 이산화(5개 구간으로 나눔)
titanic["Age"]
titanic.Age.min() #Age열의 최솟값
titanic.Age.max() #Age열의 최댓값
bins=np.linspace(titanic.Age.min(),titanic.Age.max(),5) # 5개의 구간으로 지정
bins


titanic["Age_bin"]=np.digitize(titanic["Age"],bins) #Age_bin이라는 이산화변수를 담는 열 생성
titanic
```

```python
#1-3) Fare를 이산화(3개 구간으로 나눔)
titanic["Fare"]
bins_t=np.linspace(titanic.Fare.min(),titanic.Fare.max(),3) #3개 구간 
bins_t

titanic["Fare_bin"]=np.digitize(titanic["Fare"],bins_t)
titanic
```

```python
#1-4) embarked를 원핫인코딩

pd.get_dummies(titanic["Embarked"]) # C,Q,S 범주에 대해 이항변수화
```

```python
4.
bins = [1, 15, 25, 35, 60, 99]
labels = ["미성년자", "청년", "중년", "장년", "노년"]
타이타닉호 승객을 사망자와 생존자 그룹으로 나누고 각 그룹에 대해 '미성년자', '청년', '중년', '장년', '노년' 승객의 비율을 구한다. 
각 그룹 별로 비율의 전체 합은 1이 되어야 한다.
```

```python
titanic=pd.read_csv("train .csv")

bins = [1, 15, 25, 35, 60, 99]
labels = ["미성년자", "청년", "중년", "장년", "노년"]

titanic["age_cat"]=pd.cut(titanic["Age"],bins,labels=labels) #age를 labels로 분류한 값을 가진 age_cat 생성

titanic["live"]=np.where(titanic["Survived"]==1,"live",None)  #생존자(1)은 live
titanic["dead"]=np.where(titanic["Survived"]==0,"dead",None)   #생존자(0)은 dead

dead_sum=titanic.groupby("dead")["age_cat"].value_counts().sum()   #생존자(0)에 대한 총 합
df_dead=titanic.groupby("dead")["age_cat"].value_counts()/dead_sum #사망자 퍼센테이지 : 각 age_cat의 갯수 / 생존자(0)의 총 합
df_dead
```

```python
live_sum=titanic.groupby("live")["age_cat"].value_counts().sum()  #생존자(1)에 대한 총 합
df_live=titanic.groupby("live")["age_cat"].value_counts()/live_sum  #생존자 퍼센테이지 : 각 age_cat의 갯수 / 생존자(1)의 총 합
df_live
```

```python
res=pd.DataFrame([df_dead, df_live], index=["dead","live"]) 
res
```
