```python
import pandas as pd
import numpy as np

pieces=[]
for year in range(1880,2011):
    path=("names/names/yob%d.txt" %year)
    df=pd.read_csv(path, names=["name","sex","births"])
    df["year"]=year
    pieces.append(df)
names=pd.concat(pieces, ignore_index=True)
names
```

```python
names["name"].value_counts()

#남자아이
df_male=names[names["sex"]=="M"]
df_male.set_index("sex", inplace=True)
m_2010=df_male[df_male["year"]==2010] #2010년
m_1950=df_male[df_male["year"]==1950] #1950년
m_1900=df_male[df_male["year"]==1900] #1900년


#여자아이
df_female=names[names["sex"]=="F"]
df_female.set_index("sex", inplace=True)
f_2010=df_female[df_female["year"]==2010] #2010년
f_1950=df_female[df_female["year"]==1950] #1950년
f_1900=df_female[df_female["year"]==1900] #1900년
```
